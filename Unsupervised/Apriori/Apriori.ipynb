```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv",header = None)
```

```python
dataset
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
    
#we have to structure like this because Apriori algo  expecting this as input
```

```python
transactions
```

```python
# Training Apriori on the dataset
from apyori import apriori
```

```python
help(apriori)
```

**Min_Support**  
purchased 3 times a day for 7 days = 3*7 = 21  
total transaction = 7500  
Support = 21/7500 = .0028    

**Lift**  
High Lift good rules  

**Varies with different buissness problems and do trial and error **




```python
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2) 
# making rules for two items
```

```python
# Visualising the results
results = list(rules)
```
